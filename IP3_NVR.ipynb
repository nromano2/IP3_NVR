**Function to adjust Date Column to datetime and insert a Company Name column**

```python
def dataset(filename):
    df = pd.read_csv(f"{filename}.csv")
    df['Date'] = pd.to_datetime(df['Date'], format="%d/%m/%Y")
    df.insert(0, 'Company', filename)
    return df
```

**Daily view Function**   
Created a function that returns an area graph that displays data of a specific company on a specific date

```python
def daily_view(dataframe, company, date, color):
    figure3 = plt.figure()
    dataframe.query(f"Date == '{date}' and Company == '{company}'").iloc[:,2:].T.plot(kind='area', color = color)
    plt.gca().xaxis.set_major_locator(plt.MaxNLocator(6))
    plt.title(f'Power Production: {company}/{date}')
    plt.xlabel('Time of day')
    plt.ylabel('TW/hr Generated')
    return figure3
```

```python
daily_view(df, '50Hertz', '2019-08-24', 'purple')
```

**Heat Map**

```python
df_taxis.head()
```

```python
#Created another taxis dataframe consisting of only the number of passengers picked up and at what day and hour in March
df_taxis2 = df_taxis.passengers.groupby([df_taxis['pickup'].dt.day, df_taxis['pickup'].dt.hour]).sum()
df_taxis2 = pd.DataFrame(df_taxis2)

#index consisted of day and hour column data both names pickup, renamed each column accordingly
#reset index to add day and hour to columns
df_taxis2.rename_axis(["day", "hour"], inplace=True)
df_taxis2 = df_taxis2.reset_index()
df_taxis2
```

```python
#Created pivot table to create a heatmap of the data
taxis_pivot = df_taxis2.pivot("day", "hour", "passengers")
taxis_pivot
```

```python
figure7 = plt.figure(figsize = (15,9))
sns.heatmap(taxis_pivot, cmap='Reds', linewidths=1, linecolor='black',xticklabels=['12a', '1a', '2a', '3a', '4a', 
                                                                                   '5a', '6a', '7a', '8a', '9a', 
                                                                                   '10a', '11a', '12p', '1p', '2p', '3p', 
                                                                                   '4p', '5p', '6p', '7p', '8p', '9p', 
                                                                                   '10p', '11p'])
plt.title('Passenger Numbers By Hour and Day in March')
```
